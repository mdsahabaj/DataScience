```python
#dropping irrelevant columns
data=data.drop(['EmployeeCount','EmployeeNumber'],axis=1)
```

```python
#creating dummies for attrition,changing yes no to 0 and 1
attrition_dummies=pd.get_dummies(data['Attrition']) #It converts categorical data into dummy or indicator variables.
attrition_dummies.head()
```

```python
#concat the both dataframe in columnwise
data=pd.concat([data,attrition_dummies],axis=1)
data.head()
```

```python
#drop any one column from yes and no because using one column is sufficient
data=data.drop(['No'],axis=1)
data.head()
```

```python
#dropping some least corelation columns
data=data.drop(['Age','JobLevel'],axis=1)
```

## Data Preprossing


```python
# #converting string data into integer for model feeding
# from sklearn.preprocessing import LabelEncoder
# for column in data.columns:
#     if data[column].dtype==np.number:
#         continue
#     else:
#         data[column]=LabelEncoder().fit_transform(data[column])
from sklearn.preprocessing import LabelEncoder
for column in data.columns:
    if data[column].dtypes==object:
        data[column]=LabelEncoder().fit_transform(data[column])
    else:
        continue
        
```

```python
data.head()
```

# Model Building

```python
#importing libraies
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier(n_estimators=10,criterion='entropy',random_state=0)
```

```python
x=data.drop(['Yes'],axis=1)
y=data['Yes']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=0)
```

```python
x_train.head()
```

```python
y_train.head()
```

```python
rf.fit(x_train,y_train)
```

```python
rf.score(x_train,y_train)
```

## predicting for x_test

```python
pred=rf.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,pred)
```
